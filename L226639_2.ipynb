```python
# Task 2
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
psx_url = "https://www.psx.com.pk/market-summary/"
page = requests.get(psx_url)
```

```python
if page.status_code != 200:
    raise Exception(f"Error: Unable to fetch data. HTTP Status Code: {page.status_code}")

soup = BeautifulSoup(page.text, "html.parser")
```

```python
market_tables = soup.find_all("table")
```

```python
sector_dataframes = {}
```

```python
for tbl in market_tables:
    sector_heading = tbl.find("h4")
    sector_title = sector_heading.get_text(strip=True) if sector_heading else "Unknown Sector"

    records = []
    for row in tbl.find_all("tr"):
        values = [col.get_text(strip=True) for col in row.find_all("td")]

        if len(values) == 8 and values[0] != "SCRIP":
            records.append(values)

    if not records:
        continue

    col_names = ["Scrip", "LDCP", "Open", "High", "Low", "Current", "Change", "Volume"]

    sector_df = pd.DataFrame(records, columns=col_names)
    sector_dataframes[sector_title] = sector_df

    print(f"\n Sector: {sector_title}")
    print(sector_df.head().to_string(index=False))
```
